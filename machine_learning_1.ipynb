### 데이터 병합 및 저장

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight

# 리스트 내포
fish_data = [[l,w] for l,w in zip(length, weight)]
```

머신러닝에서는 찾고자 하는 대상을 1로 두고, 아닌 대상을 0으로 둔다.
따라서 위의 데이터 중에서 bream(도미)을 1로 두고 smelt(빙어)를 0으로 맞춰주는 데이터를 입력한다.

```python
fish_target = [1] * 35 + [0] * 14

fish_target
```

###사이킷런의 기초인 KNeighbor을 사용한다.



```python
from sklearn.neighbors import KNeighborsClassifier
# import한 KNeighborClassifier의 객채를 만든다.
kn = KNeighborsClassifier()
```

fit 메서드는 주어진 데이터로 알고리즘을 훈련한다.

---


score은 학습 결과가 얼마나 정확한지를 0에서 1사이의 값으로 반환해준다.

```python
kn.fit(fish_data, fish_target)

kn.score(fish_data, fish_target)
```

###k-최근접 알고리즘
predict를 사용해서 최근접한 데이터값을 알려준다.


```python
kn.predict([[30, 600]])
```

array([1])은 bream으로 인식한 값

```python
print(kn._fit_X) # kn fit 안에 있는 x값 확인
print(kn._y) # kn fit y안에 있는 값 확인
```

sklearn.neighbors는 가까운 데이터를 참고하는데 기본값은 5로 가까운 5개의 데이터를 참고하여 결과를 도출한다.

```python
#가까운값 neighbors를 전체 데이터 갯수인 49로 설정하면 도미가 다수이므로 모두 도미로 나온다.
kn49 = KNeighborsClassifier(n_neighbors = 49) # neighbors를 49로 가지는 KNeighborsClassifier 객체 생성
kn49.fit(fish_data, fish_target)
print(kn49.score(fish_data, fish_target))

#smelt에 가까운 데이터 예측
kn49.predict([[9, 6]])
```

위 결과는 score에서 전체 데이터중에 도미의 데이터만 맞기떄문에 정확도를 계산한다면 전체 49개의 데이터중에서 35개의 도미 데이터만 정확하므로
score은 35/49 = 0.71428...으로 나온다.

### 예제

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  # k-최근접 게수 설정
  kn.n_neighbors = n

  # 점수 계산
  score = kn.score(fish_data, fish_target)
  if score < 1:
    print(n, score)
    break

```

위 결과를 통해서 참고하는 neighbors의 개수가 18개가 되면 정확도가 떨어지는것을 알 수 있다.

###전처리를 포함한 훈련 및 결과

####표준점수를 이용한 scale 조정

가장 널리 사용하는 전처리 방법 중 하나는 **표준점수**(standard score)이다. 혹은 Z점수라고도 불리는데,

표준점수는 각 특성값이 평균에서 표준편차의 몇 배 만큼 떨어져 있는지를 나타낸다. 이를 통해 실제 특성값의 크기와 상관없이 동일한 조건으로 비교 가능하다.

분산은 데이터에서 평균을 뺀 값을 모두 제곱한 다음 평균을 내어서 구한다. 표준편차는 분산의 제곱근으로 데이터가 분산된 정도를 나타낸다. 표준점수는 각 데이터가 원점에서 몇 표준편차만큼 떨어져 있는지를 나타내는 값이다.

```python
# 평균(mean)과 분산(std)를 나타내기

mean = np.mean(train_input, axis = 0)

std = np.std(train_input, axis = 0)

#평균과 분산 출력
print(mean, std)

#원본 데이터에서 평균을 빼고 표준편차로 나눠서 표준점수로 변환
scaled_data = (fish_data - mean) / std
scaled_data

# result에 따른 시각화
import matplotlib.pyplot as plt

plt.scatter(scaled_data[:, 0], scaled_data[:, 1])
plt.scatter(25, 150, marker = "^")
plt.xlabel('length')
plt.ylabel('weight')
```

위 결과에서 마커를 표준점수화를 하지 않았기 때문에 너무 높게 나온다.

```python
plt.scatter(scaled_data[:, 0], scaled_data[:, 1])

mark = ([25, 150]  - mean) / std
mark
plt.scatter(mark[0],mark[1], marker = "^")
plt.xlabel('length')
plt.ylabel('weight')
```

#### 전처리 후 데이터 정리 및 시각화

scale을 조정한 데이터를 다시 나누고 결과 확인하기

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split


# scale을 조정한 data를 train과 test 데이터로 나누기
train_input, test_input, train_target, test_target = train_test_split(
    scaled_data, fish_target, stratify = fish_target, random_state = 42)

plt.scatter(train_input[:,0], train_input[:, 1])
plt.scatter(test_input[:,0], test_input[:, 1])
plt.scatter(mark[0], mark[1], marker = "^")
plt.xlabel("length")
plt.ylabel("weight")
plt.show()

```

### 결과 확인

```python
from sklearn.neighbors import KNeighborsClassifier


kn = KNeighborsClassifier()

print(indexes)

kn.fit(train_input, train_target)

distances, indexes =kn.kneighbors([mark])

kn.score(test_input, test_target)
kn.predict([mark])


print(indexes)
```

거리를 기반으로 결과값을 예측하는 알고리즘을 사용하는 경우 scale의 차이때문에 **표준점수**{(data - mean) / std}를 통해서 전처리를 하면 더 합당한 결과를 얻어낼 수 있다는 사실을 알게 되었다.
